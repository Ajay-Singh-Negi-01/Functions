6. Explain the concept of generators in Python and how they are defined.     
  - Generators are a special type of function that allows you to iterate over data lazily, meaning one item at a time, without storing the entire sequence in memory.   
  They are defined like normal functions but use the **yield** keyword instead of **return**.   
  Generators are useful for working with large data or when you want to save memory.    
  Key Points:   
  . Use yield to pause and resume function execution.    
  . Automatically create an iterator.   
  . More memory-efficient than lists when handling large sequences.   
  

```python
def count_up_to(n):
  i = 1
  while i<= n:
    yield i
    i += 1


#Create generator
counter = count_up_to(5)


#Iterate through generator
for num in counter:
    print(num)
```

7.  What are the advantages of using generators over regular functions?    
  - Generators offer several key benefits, especially when dealing with large datasets of streams of data. Unlike regular fuctions that return the entire resrult at once, generators yield values one at a time, which makes them more memory-efficient and faster for many use cases.   
  Advantages:     
  a. Memory Efficiency-    
    Generators don't store the entire sequence in memory - they generate each value on the fly.    
    Ideal for processing large files or infinite sequences.    
  b. Faster Execution(Initial)-    
    Since values are produced only when neeeded, execution starts faster compared to building a full list.    
  c. Simpler Code for Iterators-   
    Writing a generator with yield is simpler than implementing a class with  __iter__() and __next__().   
  d. Support for infinite Data-    
    You can generate data endlessly without crashing the system(like Fibonacci numbers).   


```python
# Generator to yield first n squares

def squares(n):
  for i in range(n):
    yield i ** 2


for num in squares(5):
  print(num)



# This generator saves memory because it doesn't store [0,1,4,9,16] all at once.
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?     
  - Python provides three powerful function tools - map(), filter(), and reduce() - to process iterables like lists in a clean and efficient way:   
  a. map(function, iterable)   
    . Applies a fiven fuction to each item in the iterable.    
    . Returns a map object (need to be converted to a list).  
  b. Filter(function , iterable)      
    . Filters items in the iterable bassed on a condition (returns only items where the function returns True).       
  c. Reduce(function, iterable) (from funclools)      
    . Reduces the iterable to a single cumulative value by repeatedly applying the function.    



```python
# Example:

#importing reduce
from functools import reduce

#original list
numbers = [1,2,3,4,5]

# 1. map(): Double each number
doubled = list(map(lambda x : x * 2, numbers))
print("map() - Doubled numbers:", doubled)

# 2. filter(): keep only even numbers
evens = list(filter(lambda x : x % 2 == 0, numbers))
print("filter() - Even numbers:", evens)

# 3. reduce(): Multiply all numbers
product = reduce(lambda x, y: x * y, numbers)
print("reduce() - Product of numbers:", product)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.      


```python
def is_prime(n):
  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):     # only check till square root of n
    if n % i == 0:
      return False
  return True


#Check and print all prime numbers from 1 to 200

for num in range(1,200):
  if is_prime(num):
    print(num, end=" ")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
# Iterator class for Fibonacci sequence
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage
n = 10  # number of Fibonacci terms
fib = FibonacciIterator(n)

for num in fib:
    print(num, end=" ")
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.    



```python
""""
Formula to Convert Celsius to Fahrenheit

Fahrenheit = (Celsius * 9/5) + 32

"""

# Define a list of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Function to convert Celsius to Fahrenheit
def to_fahrenheit(c):
    return (c * 9/5) + 32

# Use map to apply the conversion to all temperatures
fahrenheit_temps = list(map(to_fahrenheit, celsius_temps))

# Print the result
print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Input string

input_str = "Python Programming is Fun"

# A function to chech if a character is not a vowel

def is_not_vowel (char):
  return char.lower() not in ["a","e","i","o","u"]

#filter to remove vowels

no_vowels = " ".join (filter(is_not_vowel, input_str))

print("Original string:", input_str)
print("String without vowels: ", no_vowels)
```
